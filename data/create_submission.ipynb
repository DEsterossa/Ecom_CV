```python
import pandas as pd
from PIL import Image
import base64
import io
from pathlib import Path

predicted_mask_dir = Path("predicted")

rows = []
for file in predicted_mask_dir.glob("*.png"):
    with Image.open(file) as im:
        buf = io.BytesIO()
        im.save(buf, format="PNG")
        image_utf = base64.b64encode(buf.getvalue()).decode("utf-8")
        rows.append({"filename": file.stem, "image_utf": image_utf})

submission_csv_path = predicted_mask_dir / "submission.csv"
pd.DataFrame(rows).to_csv(submission_csv_path, index=False)
print(f"Saved submission CSV with encoded images to {submission_csv_path}")
```
